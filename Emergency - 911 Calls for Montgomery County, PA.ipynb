```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df['title'].nunique()
```

```python
def reason(x):
    new_str =  x.split(':')
    reason = new_str[0]
    return reason
```

```python
df['Reason'] = df['title'].apply(lambda title:title.split(':')[0])
```

```python
df['Reason']
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x=df['Reason'],data=df,palette='viridis')
```

```python
type(df['timeStamp'][0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
time = df['timeStamp'][0]
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time:time.hour)
df['Month'] = df['timeStamp'].apply(lambda time:time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time:time.day_of_week)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
sns.countplot(x=df['Day of Week'],data=df,hue=df['Reason'],palette='viridis')
#To relocate the legend
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
```

```python
sns.countplot(x=df['Month'],data=df,hue=df['Reason'],palette='viridis')
#To relocate the legend
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda t:t.date())
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour)
```
